```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#ガス増幅率について

# テキストファイルの読み込み
file_path = 'combined_data.txt'

# 特定の行のデータを読み込む（3行目から6行目までを取得）
# header=Noneでヘッダーなしとして読み込む
data = pd.read_csv(file_path, delim_whitespace=True, header=None, comment='#')

# 表示したい特定の行をスライス（例：2行目から5行目まで）
selected_data = data.iloc[0]  # 0-indexedで3行目から6行目

# 表形式で表示
print("選択したデータ:")
print(selected_data)

# 偶数インデックスを y 値、奇数インデックスを誤差に分割
y_values = selected_data[::2]  # [386.009, 256.982]
y_errors = selected_data[1::2]  # [2.861, 2.401]

# x 座標
x_values = [1460, 1500]

# グラフのプロット
plt.figure(figsize=(8, 6))
plt.errorbar(x_values, y_values, yerr=y_errors, fmt='o', label='データ', capsize=5)
plt.yscale('log')  # y軸を対数スケールに設定
plt.xlabel('Voltage(V)')
plt.ylabel('gas Amplified rate')
# plt.title('誤差込み片対数グラフ')
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
# plt.legend()
plt.show()
```
